Questão 7:
Ao prospectar clientes, é importante que sejam mapeadas características capazes de definir perfis específicos, para que se possam estabelecer campanhas mais direcionadas a cada público. Dentre estas características, considerando o contexto de ofertas de produtos para clientes PJ de um banco, pode ser importante saber, por exemplo, quanto em média o cliente PJ fatura por dia em uma semana típica.

Neste contexto, você precisa implementar um protótipo de uma solução, em Python, que faz o seguinte:

recebe, como entrada, o faturamento total de cada dia de uma semana (recebendo, portanto, 7 números);
exibe na tela, como saída, o faturamento diário médio (com base nas informações de faturamento da semana informada).

```python
# Lista com os dias da semana para facilitar na digitação dos inputs
dias_da_semana = ["domingo", "segunda-feira", "terça-feira", "quarta-feira", "quinta-feira", "sexta_feira", "sábado"]

#inicialização da lista que receberá os valores do faturamento de cada dia
faturamentos = list()

#Laço de repetição para entradas dos faturamentos de cada dia
for dia in dias_da_semana:
    #While será o responsável por garantir que o usuário entre com um valor monetário não negativo
    while True:
        faturamento = float(input(f"Entre com o valor do faturamento de {dia}: R$")) #recebe o faturamento do usuário e converte para float
        if faturamento >= 0: #apenas valores do faturamento positivos ou 0
            faturamentos.append(faturamento) #valor válido é salvo na lista
            break #como o valor é válido, sai fora do while
        print("Faturamento não pode ser negativo. Digite novamente") #caso for digitado um valor negativo

#cálculo da média dos faturamentos diários:
soma = 0 #inicializada variável responsável por armazenar a soma de todos os faturamentos
for fat in faturamentos:
    soma += fat
media = soma / len(faturamentos)

print(f"\nA média dos faturamentos da semana: R${media:.2f}")
```

Questão 8:
Trabalhando na equipe de facilities de um banco, muitas vezes você faz compras internacionais, e acaba se deparando com diversas unidades de medida que caracterizam os produtos: em alguns países, medidas como onça fluida, libra e pés são muito comuns para medir volume, massa e comprimento, respectivamente.

Neste contexto, escreva um programa que seja uma calculadora para converter as medidas mencionadas acima para o sistema internacional (litro, quilograma e metro). Os fatores de conversão são os seguintes:

1 onça fluida = 0,0295735 litros 1 libra = 0,453592 quilogramas 1 pé = 0,3048 metros

A calculadora deve ser capaz de realizar tanto a conversão para o sistema internacional quanto a conversão de volta para as unidades originais.

Seu programa deve receber os inputs adequados do usuário para que seja completamente especificada a quantidade a ser convertida, e quais as unidades envolvidas na conversão desejada. Por fim, o programa deve exibir na tela tanto o valor na unidade original quanto na unidade convertida.

Exemplo: se o usuário quiser converter 5 libras para quilogramas, a mensagem a ser exibida na tela como resultado do programa deve ser: "5.0 libras é igual a 2.26796 quilogramas.".

```python
#Imprime na tela um menu de opções
print("****Calculdadora de Conversão***")
print("V: Volume")
print("M: Massa")
print("C: Comprimento\n")

#Aceita apenas que o usário digite V para volume, M para massa ou C para comprimento
while True: 
    opcao_menu = input("Digite 'V' para Volume, 'M' para Massa ou 'C' para Comprimento: ").upper()
    if opcao_menu == 'V' or opcao_menu == 'M' or opcao_menu == 'C':
        break
    print("Digite novamente. Apenas V, M ou C")

#Caso for escolhido volume
if opcao_menu == 'V':
    while True: #Aceita apenas O para Onça e L para Litros
        opcao_volume = input("Digite 'O' para entrar com o valor em Onça Fluida ou 'L' para Litros").upper()
        if opcao_volume == 'O' or opcao_volume == 'L':
            break
        print("Digite novamente. Apenas O ou L")
    
    #Converte de onça para Litros
    if opcao_volume == 'O':
        valor_volume_onca = float(input(f"Agora digite o valor em Onça Fluida para ser convertido para Litros"))
        valor_valume_litros = valor_volume_onca * 0.0295735
        print(f"{valor_volume_onca} Onça Fluida é igual a {valor_valume_litros} Litros")
    elif opcao_volume == 'L': #Converte de litros para onça
        valor_volume_litros = float(input(f"Agora digite o valor em Litros para ser convertido para Onça Fluida"))
        valor_valume_onca = valor_volume_litros / 0.0295735
        print(f"{valor_volume_litros} Litros é igual a {valor_valume_onca} Onça Fluida")

#Caso for escolhido massa
elif opcao_menu == 'M':
    while True: #Aceita apenas L para Libras e Q para Kg
        opcao_massa = input("Digite 'L' para entrar com o valor em Libras ou 'Q' para Quilogramas").upper()
        if opcao_massa == 'L' or opcao_massa == 'Q':
            break
        print("Digite novamente. Apenas L ou Q")
    
    #converte de libras para Kg
    if opcao_massa == 'L':
        valor_massa_libras = float(input(f"Agora digite o valor em Libras para ser convertido para Quilogramas"))
        valor_massa_quilo = valor_massa_libras * 0.453592
        print(f"{valor_massa_libras} Libras é igual a {valor_massa_quilo} Quilogramas")
    elif opcao_massa == 'Q': #Converte de kg para libras
        valor_massa_quilo = float(input(f"Agora digite o valor em Quilogramas para ser convertido para Libras"))
        valor_massa_libras = valor_massa_quilo / 0.453592
        print(f"{valor_massa_quilo} Quilogramas é igual a {valor_massa_libras} Libras")

#Caso for escolhido comprimento
elif opcao_menu == 'C':
    while True: #Aceita apenas P para pés e M para metros
        opcao_comprimento = input("Digite 'P' para entrar com o valor em Pés ou 'M' para metros").upper()
        if opcao_comprimento == 'P' or opcao_comprimento == 'M':
            break
        print("Digite novamente. Apenas P ou M")
    
    #Converte de pés para metros
    if opcao_comprimento == 'P':
        valor_comprimento_pes = float(input(f"Agora digite o valor em Pés para ser convertido para Metros"))
        valor_comprimento_metros = valor_comprimento_pes * 0.3048
        print(f"{valor_comprimento_pes} Pés é igual a {valor_comprimento_metros} Metros")
    elif opcao_comprimento == 'M': #Converte de metros para pés
        valor_comprimento_metros = float(input(f"Agora digite o valor em Metros para ser convertivo para Pés"))
        valor_comprimento_pes = valor_comprimento_metros / 0.3048
        print(f"{valor_comprimento_metros} Metros é igual a {valor_comprimento_pes} Pés")



```

Questão 9:
Um conceito muito comum no mercado financeiro é o de média móvel. A ideia da média móvel é suavizar uma série temporal de dados para identificar tendências ao longo do tempo. Em vez de considerar cada ponto de dados individualmente, a média móvel calcula a média de um conjunto de dados dentro de uma janela de tempo específica, deslocando essa janela ao longo da série temporal.

Em outras palavras: ao calcular a média móvel com janela de tamanho n para uma lista de valores, cada elemento da lista original é substituído pela média dos n elementos dentro da janela que termina no próprio elemento -- ou seja, a soma dos n-1 valores anteriores ao elemento com o valor do próprio elemento, dividido por n.

Considere a lista de 10 valores a seguir: [3, 5, 8, 2, 7, 10, 3, 6, 9, 4].

Escreva um programa que calcula a média móvel em janelas de 3 valores dessa lista (ou seja, n=3) e exiba na tela a lista resultante deste procedimento.

```python
lista =  [3, 5, 8, 2, 7, 10, 3, 6, 9, 4] #criada lista do enunciado
n = 3
medias_moveis = [] #iniciada lista com os valores da média móvel

for i in range(len(lista)-n+1): #itera a lista até os últimos n elementos, no caso 3
    soma = lista[i] + lista[i+1] + lista[i+2] #soma os 3 elementos
    media = soma/3 
    medias_moveis.append(media) #adiciona a média na lista
print(medias_moveis)
```

Questão 10:
Um sistema de avaliação de saúde será desenvolvido por você para ajudar médicos do trabalho a avaliar a saúde de seus pacientes, que são funcionários de um grande banco.

Os pacientes recebem uma nota de 0 a 100 para cada um dos seguintes critérios: preparo físico, alimentação saudável, saúde do coração e saúde mental.

Um paciente com nota média acima de 70 é considerado saudável, enquanto pacientes com nota média abaixo de 70 são enviados para uma avaliação completa. Entretanto, caso a saúde do coração ou a saúde mental de um paciente estejam abaixo de 30, o paciente será encaminhado para uma avaliação completa independente da média.

Escreva um programa que implementa a lógica acima, em que é exibido na tela o resultado da avaliação ("paciente saudável" ou "paciente encaminhado para avaliação completa") com base nas notas para os 4 critérios supracitados (você pode atribuir valores específicos para as notas para testar seu programa).

```python
#lista com os critérios para facilitar na hora do input
criterios = ["preparo físico", "alimentação saudável", "saúde do coração", "saúde mental"]

notas = list() # inciada uma lista com as notas de cada critério em sequência

for criterio in criterios:
    while True: #só aceita notas entre 0 e 100
        nota = float(input(f"Digite a nota (0 a 100) do {criterio}: "))
        if nota >= 0 and nota <=100:
            notas.append(nota) #adiciona a nota na lista de notas
            break
        print("Digite apenas números de 0 a 100")

media = sum(notas)/len(criterios) # calcula a média

#informa os dados enviados
print("Notas do paciente: ")
for i in range(len(criterios)):
    print(f"{criterios[i]} = {notas[i]}")
print(f"Média = {media}")

if media > 70:
    if notas[2] < 30 or notas[3] < 30:
        print("Paciente encaminhado para avaliação completa")
    else:
        print("Paciente saudável")
else:
    print("Paciente encaminhado para avaliação completa")
```
